```python
import pandas as pd
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.getOrCreate()

tables = ["customers", "sales", "sales_orders"]

base_url = "https://raw.githubusercontent.com/Bhevendra/ML-Datasets/refs/heads/main/retail_data/"

base_path = "/Volumes/ecommerce_analytics/bronze/raw_data/"



for table in tables:

    print(f"Processing the table {table}")
    file_url = f"{base_url}{table}.csv"
    print(f"Processing the URL {file_url}")

    df = pd.read_csv(file_url)

    spark_df = spark.createDataFrame(df)

    output_path = f"{base_path}{table}/"

    spark_df.write.format("csv").mode("overwrite").option("header", "true").save(output_path)

    print(f"Writing {table} to {output_path}")


print("All tables processed")
```
