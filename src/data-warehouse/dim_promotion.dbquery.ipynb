```sql
create or replace table ecommerce_analytics.data_warehouse.dim_promotion AS
select 
distinct
cast(promo_id as int),
cast(promo_disc as double) as promo_discount,
cast(promo_item as string),
cast(promo_qty as int) as promo_quantity
from
batch_1.data.sales_orders
where promo_id is not null

```
