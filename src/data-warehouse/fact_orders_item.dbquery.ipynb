```sql
CREATE OR REPLACE TABLE ecommerce_analytics.data_warehouse.fact_orders_item AS
SELECT 
DISTINCT
cast(order_number as bigint) AS order_id,
customer_id,
product_id,
promo_id,
date_format( to_timestamp(order_datetime), 'yyyyMMdd') AS order_date_key ,
qty as quantity,
price,
qty*price AS amount
FROM batch_1.data.sales_orders



```
