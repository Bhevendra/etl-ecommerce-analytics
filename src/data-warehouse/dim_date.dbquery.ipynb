```sql
CREATE OR REPLACE TABLE ecommerce_analytics.data_warehouse.dim_date AS

SELECT
    CAST(date_format(d, 'yyyyMMdd') AS INT) AS date_key,
    d AS full_date,
    day(d) AS day_num,
    month(d) AS month_num,
    year(d) AS year_num,
    quarter(d) AS quarter_num
FROM (
    SELECT explode(
        sequence(
            to_date('2018-01-01'),
            to_date('2026-12-31'),
            interval 1 day
        )
    ) AS d
);
```
